```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import operator
from infomap import infomap
from pymongo import MongoClient
from matplotlib.colors import LogNorm
from matplotlib.gridspec import GridSpec
%matplotlib inline
```

Reading the the follow graph.

```python
filename = "/Users/arefindk/Desktop/github_merged_pull/trimmed_follow_users_strong_int.graphml"
G  = nx.read_graphml(filename)
```

```python
##create matrix for graph G
##each entry is correlation between i and j
e_ij = nx.attribute_mixing_matrix(G,attribute="lang")
r = nx.attribute_assortativity_coefficient(G,'lang')
##visualize matrix
fig = plt.figure(figsize=(16,10))
plt.title(r"Language mixing matrix of follow network, $r$="+str(r),fontsize=14)
plt.pcolormesh(e_ij, norm=LogNorm(vmax=e_ij.max()),cmap='plasma')
plt.colorbar()
plt.show()
```
