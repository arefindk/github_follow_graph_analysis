```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import operator
import infomap
from pymongo import MongoClient
from matplotlib.colors import LogNorm
from matplotlib.gridspec import GridSpec
%matplotlib inline
```

Reading the the merged-pull-request graph.

```python
filename = "/Users/arefindk/Desktop/github_merged_pull/trimmed_merged_pull_users_strong_int.graphml"
G  = nx.read_graphml(filename)
```

```python
##create matrix for graph G
##each entry is correlation between i and j
e_ij = nx.degree_mixing_matrix(G,x='in',y='in')
r = nx.degree_pearson_correlation_coefficient(G,x='in',y='in')
##visualize matrix
fig = plt.figure(figsize=(16,10))
plt.title(r"Degree Correlation Matrix of merged-pull network (in-in), $r$="+str(r),fontsize=14)
plt.pcolormesh(e_ij, norm=LogNorm(vmax=e_ij.max()),cmap='plasma')
plt.colorbar()
plt.show()
```

```python
##each entry is correlation between i and j
e_ij = nx.degree_mixing_matrix(G,x='in',y='out')
r = nx.degree_pearson_correlation_coefficient(G,x='in',y='out')
##visualize matrix
fig = plt.figure(figsize=(16,10))
plt.title(r"Degree Correlation Matrix of merged-pull network (in-out), $r$="+str(r),fontsize=14)
plt.pcolormesh(e_ij, norm=LogNorm(vmax=e_ij.max()),cmap='plasma')
plt.colorbar()
plt.show()
```
